```python
# Function to calculate age and age range
def get_age(date):
    if pd.isna(date):
        return None  # Return None for NaT
    today = datetime.now()
    return today.year - date.year - ((today.month, today.day) < (date.month, date.day))

def get_agerange(age):
    if age is None:
        return None  # Return None for NaN age
    if age < 10:
        return 'Under 10'
    elif age < 20:
        return '10-19'
    elif age < 30:
        return '20-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    elif age < 60:
        return '50-59'
    else:
        return '60 and above'
```

```python
# Calculating age and age range for analysis/reporting purposes
customer['Age'] = customer['Birthday'].apply(get_age)
customer['AgeRange'] = customer['Age'].apply(get_agerange).astype('category')
```

```python
# Removing spaces in column names for MySQL
customer.columns = customer.columns.str.replace(' ', '')
```

```python
# Renaming columns to have unique names
customer.rename(columns={
    'City'      : 'Cust_City',
    'StateCode' : 'Cust_StateCode',
    'State'     : 'Cust_State',
    'ZipCode'   : 'Cust_ZipCode',
    'Country'   : 'Cust_Country',
    'Continent' : 'Cust_Continent'
}, inplace=True)
```

```python
sales['OrderDate']=pd.to_datetime (sales['OrderDate'])
sales['DeliveryDate']=pd.to_datetime(sales['DeliveryDate'])
sales.head(50)
```

```python
# Assuming 'sales' is your DataFrame and the 'OrderDate' and 'DeliveryDate' are in datetime format
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])
sales['DeliveryDate'] = pd.to_datetime(sales['DeliveryDate'])

# 1. Fill DeliveryDate for in-store orders (StoreKey != 0)
sales.loc[sales['DeliveryDate'].isna() & (sales['StoreKey'] != 0), 'DeliveryDate'] = sales['OrderDate']

# 2. Fill DeliveryDate for online orders (StoreKey == 0) by adding 7 days to OrderDate
sales.loc[sales['DeliveryDate'].isna() & (sales['StoreKey'] == 0), 'DeliveryDate'] = sales['OrderDate'] + pd.Timedelta(days=7)

# 3. Fallback for any remaining missing values using forward fill and backward fill
sales['DeliveryDate'] = sales['DeliveryDate'].fillna(method='ffill')
sales['DeliveryDate'] = sales['DeliveryDate'].fillna(method='bfill')

```
