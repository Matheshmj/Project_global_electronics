#correlations between customer demographics (age, gender) and product

```python
cust_orderfreq = final_merged.groupby(['AgeRange', 'Gender', 'Category']).size().reset_index(name='order_freq')
cust_orderfreq_df = cust_orderfreq.reset_index()

pivot_df = cust_orderfreq_df.pivot(index=['AgeRange', 'Gender'], columns='Category', values='order_freq')
pivot_df_normalized = pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100

fig, axes = plt.subplots(1, 2, figsize=(16, 12))
for i, dataframe in enumerate([pivot_df, pivot_df_normalized]):
    sns.heatmap(dataframe, cmap='YlGnBu', annot=True, fmt='.02f', linewidths=0.5, ax=axes[i])
    if i == 0:
        axes[i].set_title("Purchase Frequency by Age and Gender")
    else:
        axes[i].set_title("Normalized Purchase Frequency by Age and Gender")
    axes[i].set_xlabel("Product Category")
    axes[i].set_ylabel("Demographics (Age Group, Gender)")
    axes[i].tick_params(axis='x', rotation=90)
```

1).Gender-Neutral Purchasing Behavior: Both males and females within the same age groups show similar buying habits across product categories. This means gender doesn't significantly influence what customers prefer to purchase.

2).Universal Product Appeal: Regardless of age or gender, customers tend to prefer the same types of products, as shown by the 'normalized purchase frequency heatmap.' This indicates that certain products, like computers, cell phones, music, movies, and audiobooks, have broad appeal across diverse customer groups.

3).Higher Purchases in Older Age Group: Customers aged 60 and above tend to purchase more frequently than younger age groups. However, this is likely due to their larger presence in the customer base, not necessarily a stronger preference for specific products compared to other age groups.

```python
daily_revenue = final_merged.groupby('OrderDate')['RevenueUSD'].sum().reset_index()

avg_revenue = daily_revenue['RevenueUSD'].median()
print(f'Average Revenue Per Day: ${avg_revenue:.2f}')

# Resample by week and calculate total revenue per month
weekly_revenue = final_merged.resample('W', on='OrderDate')['RevenueUSD'].sum().reset_index()

# Calculate the average revenue per week
avg_weekly_revenue = weekly_revenue['RevenueUSD'].median()
print(f'Average Revenue Per Week: ${avg_weekly_revenue:.2f}')

# Resample by month and calculate total revenue per month
monthly_revenue = final_merged.resample('ME', on='OrderDate')['RevenueUSD'].sum().reset_index()

# Calculate the average revenue per month, taking median as it may have outliers
avg_monthly_revenue = monthly_revenue['RevenueUSD'].median()
print(f'Average Revenue Per Month: ${avg_monthly_revenue:.2f}')

# Resample by year and calculate total revenue per year
yearly_revenue = final_merged.resample('YE', on='OrderDate')['RevenueUSD'].sum().reset_index()

# Calculate the average revenue per year
avg_yearly_revenue = yearly_revenue['RevenueUSD'].mean()
print(f'Average Revenue Per Year: ${avg_yearly_revenue:.2f}')
```

```python
import pandas as pd
import pymysql
import sqlalchemy
from sqlalchemy import create_engine, text

# Database connection details
host = '127.0.0.1'
user = 'root'
password = '848981'
database = 'global_electronics'

# Create SQLAlchemy Engine 
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')
```

```python

# Create the Customers table (if it doesn't exist)
create_customer = """
CREATE TABLE IF NOT EXISTS customers (
    CustomerKey INT PRIMARY KEY NOT NULL,
    Gender VARCHAR(10),
    Name VARCHAR(255),
    City VARCHAR(255),
    StateCode VARCHAR(255),
    State VARCHAR(255),
    ZipCode VARCHAR(20),
    Country VARCHAR(100),
    Continent VARCHAR(100),
    Birthday DATE,
    Age INT,
    AgeRange VARCHAR(20)
)"""

# Use SQLAlchemy to execute the table creation
with engine.connect() as connection:
    connection.execute(text(create_customer))
    
# Push the Customer data into MySQL
customer.to_sql('customers', con=engine, if_exists='replace', index=False)

print("Data inserted successfully!")
```

```python

# SQL Query to Create Product Table
create_product = """
CREATE TABLE IF NOT EXISTS product (
    ProductKey INT PRIMARY KEY AUTO_INCREMENT,
    ProductName VARCHAR(100),
    Brand VARCHAR(100),
    Color VARCHAR(50),
    UnitCostUSD DECIMAL(10, 2),
    UnitPriceUSD DECIMAL(10, 2),
    SubcategoryKey INT,
    Subcategory VARCHAR(100),
    CategoryKey INT,
    Category VARCHAR(100)
);
"""

# Execute the table creation query
with engine.connect() as connection:
    connection.execute(text(create_product)) 
    
# Push the Customer data into MySQL
product.to_sql('product', con=engine, if_exists='replace', index=False)
print("Table 'product' created successfully.")
print("Data inserted successfully into 'product' table.")
```

```python
# SQL query to create the sales table
create_sales = """
CREATE TABLE IF NOT EXISTS sales (
    OrderNumber INT,
    LineItem INT,
    OrderDate DATE,
    DeliveryDate DATE,
    CustomerKey INT,
    StoreKey INT,
    ProductKey INT,
    Quantity INT,
    CurrencyCode VARCHAR(10),
    PRIMARY KEY (OrderNumber, LineItem)
);
"""
# Execute the table creation query
with engine.connect() as connection:
    connection.execute(text(create_sales)) 
    
# Push the Customer data into MySQL
sales.to_sql('sales', con=engine, if_exists='replace', index=False)
print("Table 'sales' created successfully.")
print("Data inserted successfully into 'sales' table.")
```

```python
# SQL query to create the store table
create_store = """
CREATE TABLE IF NOT EXISTS store (
    StoreKey INT PRIMARY KEY AUTO_INCREMENT,
    Country VARCHAR(100),
    State VARCHAR(100),
    SquareMeters INT,
    OpenDate DATE,
    storeAge INT,
    OpenMonth VARCHAR(20)
);
"""

# Execute the table creation query
with engine.connect() as connection:
    connection.execute(text(create_store)) 
    
# Push the Customer data into MySQL
store.to_sql('store', con=engine, if_exists='replace', index=False)
print("Table 'store' created successfully.")
print("Data inserted successfully into 'store' table.")
```

```python
# SQL query to create the exchange table
create_exchange = """
CREATE TABLE IF NOT EXISTS exchange (
    Date DATE,
    Currency VARCHAR(10),
    Exchange DECIMAL(10, 4)
);
"""

# Execute the table creation query
with engine.connect() as connection:
    connection.execute(text(create_exchange)) 
    
# Push the Customer data into MySQL
exchange.to_sql('exchange', con=engine, if_exists='replace', index=False)
print("Table 'exchange' created successfully.")
print("Data inserted successfully into 'exchange' table.")
```
